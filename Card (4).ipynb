```python
from sklearn.preprocessing import LabelEncoder

dataClean = cardData

# fema w7id type categoriel ama numerique A2 w A14
dataClean['A2'] = dataClean['A2'].replace('?', pd.NA)
dataClean['A2'] = pd.to_numeric(dataClean['A2'])
dataClean['A14'] = dataClean['A14'].replace('?', pd.NA)
dataClean['A14'] = pd.to_numeric(dataClean['A14'])

dataClean
```

```python
# Replace missing values in categorical columns with mode
dataClean = cardData
categorical_columns = dataClean.select_dtypes(include=['object']).columns

for column in categorical_columns:
    dataClean[column] = dataClean[column].replace('?', pd.NA)
    dataClean[column].fillna(dataClean[column].mode()[0], inplace=True)
dataClean
```

```python


categorical_columns = dataClean.select_dtypes(include=['object']).columns
for column in categorical_columns:
    unique_values = dataClean[column].unique()
    print(f"Unique values in {column}: {unique_values}")
```

```python

categorical_columns = dataClean.select_dtypes(include=['object']).columns
# Apply Label Encoding to ordinal variables
label_encoder = LabelEncoder()
for column in categorical_columns:
    dataClean[column] = label_encoder.fit_transform(dataClean[column])

dataClean
```

```python
missing_values = dataClean.isnull().sum()

# Print the count of missing values for each column
print("Missing values in each column:")
print(missing_values)
```

```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# Step 1: Visual inspection
# Example: Box plot for a specific feature
dataClean['A2'].plot(kind='box')
plt.show()

# Step 2: Statistical methods
# Z-score method
z_scores = (dataClean['A2'] - dataClean['A2'].mean()) / dataClean['A2'].std()
outliers_zscore = dataClean[np.abs(z_scores) > 3]

# Interquartile range (IQR) method
Q1 = dataClean['A2'].quantile(0.25)
Q3 = dataClean['A2'].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = dataClean[(dataClean['A2'] < Q1 - 1.5 * IQR) | (dataClean['A2'] > Q3 + 1.5 * IQR)]

# Step 3: Machine learning algorithms
# Isolation Forest method
clf = IsolationForest(contamination=0.05)  # 5% contamination
clf.fit(dataClean[['A2']])
outliers_iforest = dataClean[clf.predict(dataClean[['A2']]) == -1]

# Step 4: Handling outliers
# Example: Remove outliers
dataClean['A2'] = dataClean['A2'].drop(outliers_zscore.index)
dataClean['A2'].plot(kind='box')
plt.show()
```

```python
dataClean['A1'].plot(kind='box')
plt.show()
upper_limit = dataClean['A1'].mean() + 3*dataClean['A1'].std()
lower_limit = dataClean['A1'].mean() - 3*dataClean['A1'].std()
#capping - change the outlier values to upper (or) lower limit values 
new_def = dataClean.copy()
new_def.loc [(new_def['A1']>upper_limit),'A1'] = upper_limit
new_def.loc [(new_def['A1']<lower_limit),'A1'] = lower_limit
new_def['A1'].plot(kind='box')
plt.show()
```
